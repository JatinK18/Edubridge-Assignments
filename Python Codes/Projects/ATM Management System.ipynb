# ATM Management System

```python
from time import sleep                                                           # importing module to delay statements
import datetime                                                                  # importing module to use current date-time

l=['1234']                                                                       # List of ATM PIN
class ATM():                                                                     # Creating Class with name ATM
    def __init__(self, balance = 10000):                                         # Constructor
        self.balance = balance                                                   # self.balance to give default account balance
         
    def withdraw(self,amount):                                                   # Creating a function for withdrawal
        self.amount=amount
        if self.amount<=self.balance:
            self.balance=self.balance-self.amount
            print('Please wait your transaction is under process.....')
            sleep(2)
            print('|--------------------------|')
            print('  Withdrawal Amount: ',self.amount)
            print('  Available Balance: ',self.balance)
            print('|--------------------------|')
        else:
            print('Insufficient balance')
        
    def credit(self,amount):                                                      # Creating a function for credit
        self.amount=amount
        self.balance=self.balance+self.amount
        print('..................................................')
        print('Please put the cash in the machine')
        sleep(2)
        print('Please wait your transaction is under process.....')
        sleep(2)
        print('Amount Credited Successfully\n')
        print('|----------------------------|')
        print('  Credited Amount:    ',self.amount)
        print('  Available balance = ',self.balance)
        print('|----------------------------|')
        
    
    def mini(self):                                                                 # Creating a function for mini statement
        print('|*******************************|')
        print('|--------Date--------Time-------|')
        print('|',datetime.datetime.now(),'   |')
        print('|Account number: 66600201006852 |')
        print('|Last Transaction amount: ',self.amount,'|')
        print('|Available balance: ',self.balance,'     |')
        print('|...............................|')
    
    def change(self,pi):                                                            # Creating a function for changing PIN
        print('*************************')
        o=input('Enter Your Current PIN: ') 
        if o==pi:
            n=input('Enter New PIN: ')  
            m=input('Confirm New PIN: ') 
            print('*************************')
            if n==m:
                l.pop()
                l.append(n)
                print('PIN Changed Successfully')
            else:
                print('PIN Not Matched')
        else:
            print('Current PIN Not Matched With System')
        
atm=ATM()                                                                             # Creating an object for class 'ATM()'
def call():                                                                           # creating a function for infinite loop
    print('''
                    ****WELCOME TO ATM****
                 ---PLEASE ENTER YOUR CARD---
                    ''')
    sleep(3)                                                                          # Calling function sleep() for delay
    done=False 
    while done==False:                                                                # infinite loop
        
        pi=input('Enter your PIN: ')
        if pi in l:
            print()
            print("""
                    =========MENU=========
                    |  1. Withdrawal     |
                    |  2. Credit         |
                    |  3. Mini Statement |
                    |  4. Change PIN     |
                    |  5. Exit           |
                    **********************
                  """)
            try:                                                                   # Exception handling--to try given conditions
                choice=int(input("Select an option from the Menu:  "))
                if choice==1:
                    print('\nWITHDRAWAL')
                    amount=int(input('Enter the amount: ')) 
                    atm.withdraw(amount)
                elif choice==2:
                    print('\nCREDIT')
                    amount=int(input('Enter the amount: '))  
                    atm.credit(amount)
                elif choice==3:
                    print('\nMINI STATEMENT')
                    atm.mini()
                elif choice==4:
                    print('\nCHANGE PIN')
                    atm.change(pi)
                elif choice==5:
                    sleep(2)
                    print('\nExiting...')
                    print(                  '''
                 -------------  .     .        /\       |\     |  |  /   ,======     
                       |        |     |       /  \      | \    |  | /    |
                       |        |     |      /    \     |  \   |  |/     |
                       |        |=====|     /------\    |   \  |  |\     '=====,
                       |        |     |    /        \   |    \ |  | \          |
                       |        |     |   /          \  |     \|  |  \   ======'
           
                            ''')
                    break
                else:
                    print('Enter Proper choice')
                    break
                            
            except:                                                                 # Exception handling if any error occured
                    print("Please enter proper choice.")
        else:
            print('Wrong PIN Entered')
            
call()
```
